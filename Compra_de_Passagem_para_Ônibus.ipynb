```python
def principal():
  
  matriz = [[0]*16 for _ in range(3)]

  return matriz

def venderPassagem(matriz):
  
  res = True
  v = True
  while res == True: 
    while v == True:
      linha = int(input("Digite a linha: "))
      coluna = int(input("Digite a coluna:"))
      for i in range(len(matriz)):
        for j in range(len(matriz[i])):
          if i == linha-1 and j == coluna-1:
            if matriz[i][j] != 0:
              print("Assento Ocupado!")
              v = True
            else:
              v = False
    if (linha < 1 or coluna < 1) or (linha > 3 or coluna > 16):
      print("Posição inválida!") 
    else:
      idadeNumero = int(input("Diga sua idade: 0 - 100: "))
      if idadeNumero < 0 or idadeNumero > 100:
        print("Resposta inválida!")
      else:
        if idadeNumero <= 12:
          idade = 'CR'
          res = False
        elif idadeNumero <= 18:
          idade = 'ADL'
          res = False     
        else:
          idade = 'ADU'
        res = False
  return linha,coluna,idade,idadeNumero

def ocuparPosicao(matriz,linha,coluna,idade):
  
  for i in range(len(matriz)):
    for j in range(len(matriz[i])):
      if i == linha-1 and j == coluna-1:
        matriz[i][j] = idade
          
        return matriz

def calcularPreco(linha,idade,valor):
  
  if linha == 1 or linha == 3:
      valor = (150 * (idade == 'CR')) + (260 * (idade == 'ADL')) + (370 * (idade == 'ADU'))
  else:
      valor = (100 * (idade == 'CR')) + (200 * (idade == 'ADL')) + (300 * (idade == 'ADU'))  

  return valor
      
def quantidadeAssentos(matriz):
  
  contLivre = 0
  contOcupado = []
  for i in range(len(matriz)):
    cont = 0
    for j in range(len(matriz[i])):
      if matriz[i][j] != 0:
        cont = cont + 1
      else:
        contLivre = contLivre + 1
    contOcupado.append(cont)

  return contLivre,contOcupado

def verificarPluralidade(posicao,posicao2):
    
    posicao = 'assento' + 's' * (posicao > 1)
    posicao2 = 'assento' + 's' * (posicao2 > 1)

    return posicao,posicao2
  
def mediaIdade(listaIdades):
  contCR = 0
  contADL = 0
  contADU = 0
  for i in range(len(listaIdades)):
    if 'CR' in listaIdades[i]:
      contCR = contCR + 1 
    elif 'ADL' in listaIdades[i]:
      contADL = contADL + 1 
    elif 'ADU' in listaIdades[i]:
      contADU = contADU + 1
  maior = contCR
  faixaEtaria = 'Criança'
  if contADL > maior:
    maior = contADL
    faixaEtaria = 'Adolescente'
  elif contADU > maior:
    maior = contADU
    faixaEtaria = 'Adulto'
  elif contADL == maior:
    faixaEtaria = 'Nenhum'
  elif contADU == maior:
    faixaEtaria = 'Nenhum'
  
  return faixaEtaria

def maiorIdade(listaIdade):
  maior = 0
  for i in listaIdade:
    if i > maior:
      maior = i
  return maior

def main():
  
  matriz = principal()
  res = 's'
  valor = 0
  total = 0
  lista = []
  mediaIdadeLista = []
  maiorIdadeLista = []
  
  while res.lower() == 's':
    
    inf = venderPassagem(matriz)
    posicao = ocuparPosicao(matriz,inf[0],inf[1],inf[2])
    preco = calcularPreco(inf[0],inf[2],valor)
    total = total + preco
    maiorIdadeLista.append(inf[3])
    maior= maiorIdade(maiorIdadeLista)
    assentos = quantidadeAssentos(matriz)
    mediaIdadeLista.append(inf[2])
    fxEtaria = mediaIdade(mediaIdadeLista)
    plural = verificarPluralidade(assentos[1][0],assentos[1][2])
    
    print(f"Na primeira fileira ao lado da janela, já foi ocupado {assentos[1][0]} {plural[0]}, na fileira do meio, {assentos[1][1]}, e na fileira da outra janela está ocupado mais {assentos[1][2]} {plural[1]}.") 
    print(f"Restam mais {assentos[0]} assentos livres!")
    print(f"A faixa etária predominante no momento é: {fxEtaria}, e a maior idade no ônibus é de {maior} anos")

    res = 't'

    while res != 's' and res != 'n':
      res = str(input(f"Deseja comprar mais uma vaga?, no momento o total a pagar é de: R${total} S- N-"))
      if res.lower() != 's' and res.lower() != 'n':
        print("Digite corretamente a resposta!")
      


  for i in matriz:
    for j in i:
      print(j, end=' ')
    print('\n')

main()
```
